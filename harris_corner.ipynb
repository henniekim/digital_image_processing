```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Harris Corner Detector

```python
img = cv2.imread('image/Lena.png')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
plt.title('gray image')
plt.show()
```

```python
flat = np.random.normal(loc = 50, scale=10, size=(100,100))
edge = np.random.normal(loc = 100, scale=10, size=(100,100))
edge[:,0:50] *= 2
corner = np.random.normal(loc = 100, scale = 10, size=(100,100))
corner[50:100,:] *= 2
corner[0:50,50:100] *= 2
```

```python
fig=plt.figure(figsize=(16,16))

plt.subplot(1, 3, 1)
plt.xticks(())
plt.yticks(())
plt.title('flat')
plt.imshow(flat, cmap='gray')

plt.subplot(1, 3, 2)
plt.xticks(())
plt.yticks(())
plt.title('edge')
plt.imshow(edge, cmap='gray')

plt.subplot(1, 3, 3)
plt.xticks(())
plt.yticks(())
plt.title('corner')
plt.imshow(corner, cmap='gray')
```

```python
flat_x_der = np.zeros((100,100))
flat_y_der = np.zeros((100,100))

edge_x_der = np.zeros((100,100))
edge_y_der = np.zeros((100,100))

corner_x_der = np.zeros((100,100))
corner_y_der = np.zeros((100,100))
```

```python
epsilon = 10

for i in range (epsilon,100-epsilon):
    flat_x_der
    
    flat_x_der[:,i] = flat[:,i-epsilon] - flat[:,i+epsilon]
    flat_y_der[i,:] = flat[i-epsilon,:] - flat[i+epsilon,:]
    
    edge_x_der[:,i] = edge[:,i-epsilon] - edge[:,i+epsilon]
    edge_y_der[i,:] = edge[i-epsilon,:] - edge[i+epsilon,:]
    
    corner_x_der[:,i] = corner[:,i-epsilon] - corner[:,i+epsilon]
    corner_y_der[i,:] = corner[i-epsilon,:] - corner[i+epsilon,:]
    
```

```python
fig=plt.figure(figsize=(16,8))

plt.subplot(2,3, 1)
plt.title('X derivative of flat')
plt.imshow(flat_x_der, cmap='gray')

plt.subplot(2,3,4)
plt.title('Y derivative of flat')
plt.imshow(flat_y_der, cmap ='gray')

plt.subplot(2,3, 2)
plt.title('X derivative of edge')
plt.imshow(edge_x_der, cmap='gray')

plt.subplot(2,3,5)
plt.title('Y derivative of edge')
plt.imshow(edge_y_der, cmap ='gray')

plt.subplot(2,3, 3)
plt.title('X derivative of corner')
plt.imshow(corner_x_der, cmap='gray')

plt.subplot(2,3,6)
plt.title('Y derivative of corner')

plt.imshow(corner_y_der, cmap ='gray')
```

```python
fig=plt.figure(figsize=(16,5))

plt.subplot(1,3,1)
plt.scatter(flat_x_der[:,:], flat_y_der[:,:])
plt.axis([-200,200, 200, -200])
plt.title('flat region')
#plt.show()

plt.subplot(1,3,2)
plt.scatter(edge_x_der[:,:], edge_y_der[:,:])
plt.axis([-200,200, 200, -200])
plt.title('edge region')
#plt.show()

plt.subplot(1,3,3)
plt.scatter(corner_x_der[:,:], corner_y_der[:,:])
plt.axis([-200,200, 200, -200])
plt.title('corner region')
plt.show()
```
